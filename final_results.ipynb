```python
import numpy as np

results = np.load("submission1/group_56_catch_rewards_0.npy", allow_pickle=True)
print(results)

from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline

x = np.arange(1, len(results)*10, 10) 
y = results

plt.title("Winrate over time") 
plt.xlabel("Episodes") 
plt.ylabel("Winrate") 
plt.ylim(0, 1)
plt.plot(x,y) 
plt.show()
```
